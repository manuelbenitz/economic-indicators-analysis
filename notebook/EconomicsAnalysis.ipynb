```python
# Create Schema for Spark DataFrame
def cschema(column_names):
    fields = []
    for col_name in column_names:
      #check if column name matches the pattern for years
        if re.match(r'^\d{4} \[YR\d{4}\]$', col_name):
            fields.append(StructField(col_name, DoubleType(), True)) # assing numerical type
        else:
            fields.append(StructField(col_name, StringType(), True))#assing string type
    return StructType(fields)

# Clean and rename columns that contain year values
def cyears(data):
    year_columns = [col for col in data.columns if re.match(r'^\d{4}( \[YR\d{4}\])?$', col)]
    for col_name in year_columns:
        #remove unnecessary parts of the year format for a clean year value
        clean_year = re.sub(r' \[YR\d{4}\]', '', col_name)
        #rename columns with the new format
        data = data.withColumnRenamed(col_name, clean_year)
    return data

# dictionary to hold DataFrames for different datasets
DataFrames = {}

# Creating a DataFrame for each dataset
for file_path, name in FILE_PATHS:

    dataheader = spark.read.csv(file_path, header=True, inferSchema=False).limit(1)
     #generate schema for the data based on columns
    schema = cschema(dataheader.columns)
    data = spark.read.csv(file_path, header=True, schema=schema)  # Data with schema
    data = cyears(data)  # Year processing
    data = data.withColumn("SERIES CATEGORY", lit(name.upper()))
    DataFrames[name] = data  # Store the updated DataFrame
```

```python
# create an "id" for cosmos
def cosmos_id(data, id_cols=("CountryCode","SeriesCode"), separator="_"):
  #concatenate columns with a unique ID for each record
    return data.withColumn("id", F.concat_ws(separator, *[F.col(c) for c in id_cols]))

# convert spark to pandas DataFrame
def convert_topd(sparkdata):
    return sparkdata.toPandas()

# connect cosmos container
def in_cosmosclient(endpoint, key, databasename, containername):
    client = CosmosClient(endpoint, key)
    database = client.get_database_client(databasename)
    return database.get_container_client(containername)

# conversion DataFrame into dictionary
def convert_r(datapd):
    return [
        {k: v.item() if hasattr(v, 'item') else v
         for k, v in record.items()}
        for record in datapd.to_dict(orient='records')
    ]

# uploading DataFrame to Container in cosmos
def upload_cosmos(container, records, max_retries=3):
    success = 0 #count succesful uploads
    errors = [] #collect errors
    for record in tqdm(records, desc="Uploading records"): # display progress
        for attempt in range(max_retries + 1): #retry upload errors
            try:
                container.upsert_item(body=record)# insert in cosmos
                success += 1
                break
            except Exception as e: #handle errors
                if attempt == max_retries:
                    errors.append({
                        "id": record.get('id', 'missing_id'),
                        "error": str(e),
                        "attempts": attempt + 1
                    })
    return success, errors# return details of succsesful and error uploads
```

```python
### UPLOAD
def main():
  # generate a unique id for each record
    dataid = cosmos_id(data)
    datapd = convert_topd(dataid)#converts de spark data with id to pandas
  # cosmos conection
    endpoint = "https://dataintensive.documents.azure.com:443/"
    key = "7XbcySCTQSmn1nWcS5bBQ7BbrNEmMU8Kkpkf3QHZbNOQNW8Fse8UQDv9p4A87k5QvfXi6IbddHk4ACDb3nqEfg=="
    container = in_cosmosclient(endpoint, key, "diss", "economics3")
  #convert pandas into list for cosmos inserction
    records = convert_r(datapd)
    success_count, upload_errors = upload_cosmos(container, records)
  # print sumary
    print(f"\nSuccessfully uploaded {success_count}/{len(records)} records")
    if upload_errors:
        print(f"\nFirst 5 errors:")
        for error in upload_errors[:5]:
            print(f"ID {error['id']} ({error['attempts']} attempts): {error['error']}")

if __name__ == "__main__":
    main()#execute upload process
```
